 NOTE: this crashed colab due to RAM overflow on 1.3M rows.
 fixed by processing the file in chunks of 50k rows using Counter
 and standardizing skill names to fix capitalization duplicates.


```python
import pandas as pd
from collections import Counter

# we will read the skills file in small chunks of 50000 rows at a time
# and count skills as we go without ever loading the full file into memory
chunk_size = 50000
skill_counter = Counter()

# read and process one chunk at a time
for chunk in pd.read_csv('/content/job_skills.csv', chunksize=chunk_size):

    # drop missing values in this chunk
    chunk = chunk.dropna(subset=['job_skills'])

    # split skills, explode, clean and count
    skills_in_chunk = (chunk['job_skills']
                       .str.split(',')
                       .explode()
                       .str.strip()
                       .str.lower()
                       .str.replace(r'\s+', ' ', regex=True))

    # add counts from this chunk to our running total
    skill_counter.update(skills_in_chunk.tolist())

# convert final counts to dataframe
top_skills = pd.DataFrame(skill_counter.most_common(20),
                          columns=['skill', 'total_postings'])

print(top_skills)
```

```python
import pandas as pd
from collections import Counter

chunk_size = 50000
skill_counter = Counter()

# dictionary to standardize similar skill names
# left side is what appears in data, right side is what we want it to become
skill_mapping = {
    'communication skills': 'communication',
    'customer service': 'customer service',
    'problemsolving': 'problem solving',
    'time management': 'time management',
    'attention to detail': 'attention to detail',
    'interpersonal skills': 'interpersonal skills',
    'microsoft office suite': 'microsoft office',
    'organizational skills': 'organization',
    'communication skills': 'communication',
}

for chunk in pd.read_csv('/content/job_skills.csv', chunksize=chunk_size):

    chunk = chunk.dropna(subset=['job_skills'])

    skills_in_chunk = (chunk['job_skills']
                       .str.split(',')
                       .explode()
                       .str.strip()
                       .str.lower()
                       .str.replace(r'\s+', ' ', regex=True))

    # replace similar skills with standardized names
    skills_in_chunk = skills_in_chunk.replace(skill_mapping)

    skill_counter.update(skills_in_chunk.tolist())

# convert to dataframe
top_skills = pd.DataFrame(skill_counter.most_common(20),
                          columns=['skill', 'total_postings'])

print(top_skills)
```

6 — Top 10 Countries With Most Job Postings

```python
# extract country from job_location
# job_location format is usually "City, State" or "City, Country"
# the country/state is after the last comma
jobs['country'] = jobs['job_location'].str.split(',').str[-1].str.strip()

# count postings per country
top_countries = jobs['country'].value_counts().head(10).reset_index()
top_countries.columns = ['country', 'total_postings']

print(top_countries)
```

```python
# list of us state codes
us_states = ['AL','AK','AZ','AR','CA','CO','CT','DE','FL','GA','HI','ID','IL','IN',
             'IA','KS','KY','LA','ME','MD','MA','MI','MN','MS','MO','MT','NE','NV',
             'NH','NJ','NM','NY','NC','ND','OH','OK','OR','PA','RI','SC','SD','TN',
             'TX','UT','VT','VA','WA','WV','WI','WY','DC']

# extract last part of location
jobs['last_part'] = jobs['job_location'].str.split(',').str[-1].str.strip()

# if last part is a us state code, mark country as united states
# otherwise use the last part as country
jobs['country'] = jobs['last_part'].apply(
    lambda x: 'United States' if x in us_states else x
)

# drop the helper column
jobs = jobs.drop(columns=['last_part'])

# count top 10 countries
top_countries = jobs['country'].value_counts().head(10).reset_index()
top_countries.columns = ['country', 'total_postings']

print(top_countries)
```

7 - Top Companies job aggregators removed

```python
aggregators = ['Health eCareers', 'Jobs for Humanity', 'TravelNurseSource',
               'PracticeLink', 'Energy Jobline', 'ClearanceJobs',
               'Jobot', 'Gotham Enterprises Ltd', 'VolunteerMatch',
               'DocCafe', 'ClickJobs.io', 'CyberCoders']

top_companies_clean = (jobs[~jobs['company'].isin(aggregators)]
                       ['company']
                       .value_counts()
                       .head(10)
                       .reset_index())
top_companies_clean.columns = ['company', 'total_postings']
top_companies_clean.to_csv('/content/top_companies_clean.csv', index=False)
print("top companies clean done")
print(top_companies_clean)
```

8 - Job Posting Trend Over Time

```python
# get top 5 cities
top_5_cities = ['New York', 'London', 'Houston', 'Chicago', 'Los Angeles']

# filter jobs for those cities and count job types
city_job_type = (jobs[jobs['city'].isin(top_5_cities)]
                 .groupby(['city', 'job_type'])
                 .size()
                 .reset_index())
city_job_type.columns = ['city', 'job_type', 'total_postings']
city_job_type.to_csv('/content/city_job_type.csv', index=False)
print("city job type done")
print(city_job_type)
```

9 - Remote Jobs By Country

```python
us_states = ['AL','AK','AZ','AR','CA','CO','CT','DE','FL','GA','HI','ID','IL','IN',
             'IA','KS','KY','LA','ME','MD','MA','MI','MN','MS','MO','MT','NE','NV',
             'NH','NJ','NM','NY','NC','ND','OH','OK','OR','PA','RI','SC','SD','TN',
             'TX','UT','VT','VA','WA','WV','WI','WY','DC']

jobs['last_part'] = jobs['job_location'].str.split(',').str[-1].str.strip()
jobs['country'] = jobs['last_part'].apply(
    lambda x: 'United States' if x in us_states else x)
jobs = jobs.drop(columns=['last_part'])

# filter remote jobs and remove metropolitan areas
remote_by_country = (jobs[jobs['job_type'] == 'Remote']
                     ['country']
                     .value_counts()
                     .reset_index())
remote_by_country.columns = ['country', 'remote_jobs']

# remove metropolitan area entries
remote_by_country = remote_by_country[
    ~remote_by_country['country'].str.contains('Metropolitan|Greater|Area|Metroplex',
                                                case=False)]
remote_by_country = remote_by_country.head(10)
remote_by_country.to_csv('/content/remote_by_country.csv', index=False)
print("remote by country done")
print(remote_by_country)
```

10 - Job Level vs Job Type

```python
job_level_type = (jobs.groupby(['job_level', 'job_type'])
                  .size()
                  .reset_index())
job_level_type.columns = ['job_level', 'job_type', 'total_postings']
job_level_type.to_csv('/content/job_level_type.csv', index=False)
print("job level type done")
print(job_level_type)
```
